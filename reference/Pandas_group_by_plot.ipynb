```
import pandas as pd
```

```
df = pd.read_csv("https://felixml.s3.amazonaws.com/house_prices/train.csv")
df.head()
```

```
df2=df[['SalePrice','YrSold']]
df2.dtypes
df2['YrSold'].apply(pd.to_numeric)
df2['SalePrice'].apply(pd.to_numeric)
df2.dtypes
```

```

df3 = df2.groupby(['YrSold']).mean().astype(int)
df3.plot(logy=True)
```
